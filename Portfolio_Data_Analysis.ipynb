```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df  = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
plt.figure(figsize = (10,6))
sns.histplot(data = df,x = 'price')
plt.xticks(rotation = 90)
plt.show()
```

```python
df = df[df['price']<200000]
plt.figure(figsize = (16,6))
sns.histplot(data = df,x = 'price')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (16,6))
sns.histplot(data = df,x = 'size')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.histplot(data = df,x = 'rooms')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.displot(data = df[df['rooms']<4], x = 'price',hue = 'rooms',multiple = 'stack',palette = 'Reds',height = 5,aspect =3)
plt.show()
```

```python
sns.displot(data = df[df['rooms']>4],x = 'price',col = 'rooms')
plt.show()
```

```python
sns.displot(data = df,x = 'size', height = 4,aspect = 2,kind = 'kde')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.displot(data = df,x = 'size',kde = True,height = 4,aspect =3)
plt.show()
```

```python
sns.displot(data = df[df['rooms'].isin([2,3,5])], x = 'price',hue = 'rooms',height =4,aspect = 3,kind = 'kde')
plt.show()
```

```python
plt.figure(figsize =(16,6))
sns.kdeplot(data = df[df['rooms'].isin([2,3,4])],shade = True, x = 'price',hue = 'rooms')
plt.show()
```

```python
sns.displot(data = df, x = 'size', kind = 'ecdf',height = 5,aspect =2)
plt.grid(which = 'both')
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data = df,x = 'price',kind = 'ecdf',height = 5,aspect =2)
plt.grid(which = 'both')
plt.minorticks_on()
plt.show()
```
